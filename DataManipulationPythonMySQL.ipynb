```python
# First import required libraries
import pandas as pd # use this to import the whole package
from sqlalchemy import create_engine # use this to create a part of a package
import pymysql # us this to import the entire package
```

```python
# Create an engine and connect to a database,
engine = create_engine('mysql+pymysql://xxxxx:xxxxx!@localhost/data1202')
```

```python
# create a connection string
con_1 = engine.connect()
```

```python
# Create a query to compare average global sales before and after 2005 into a dataframe df using the connection string created
df = pd.read_sql('''SELECT 
    CASE
        WHEN year < 2005 THEN 'pre-2005'
        ELSE 'post-2005'
    END AS year_category, AVG(global_sales) AS avg_global_sales
FROM
    data1202.vgsales
GROUP BY year_category
ORDER BY avg_global_sales DESC''', con_1)
```

```python
#Print the result and determine which was higher
df.head()
```

```python
# Write a query that creates a new column which labels records before 2005 as 'pre-2005' and after 2005 as 'post-2005'
# into a dataframe df1
df1 = pd.read_sql('''SELECT 
    Name,
    Platform,
    Global_Sales,
    CASE
        WHEN Year < 2005 THEN 'pre-2005'
        ELSE 'post-2005'
    END AS year_category
FROM
    vgsales''', con_1)
```

```python
#Import required libraries
import pandas as pd # use this to import the whole package
```

```python
# Create dataframe
vgsales = pd.read_csv('vgsales.csv')

#Check to be sure it's loaded correctly
vgsales.head()
```

```python
# Query to get average Global_sales where yeare is before 2005
average_before_2005 = vgsales[vgsales['Year'] < 2005]['Global_Sales'].mean()
print(average_before_2005)
```

```python
# Query to get average Global_sales where yeare is after 2005
average_after_2005 = vgsales[vgsales['Year'] > 2005]['Global_Sales'].mean()
print(average_after_2005)
```

```python
#Query to compare the global sales averages before and after 2005
if average_before_2005 > average_after_2005:
    
    print("The average of global sales was higher before 2005.")
    
else:
    
    print("The average of global sales was higher after 2005.")
```
